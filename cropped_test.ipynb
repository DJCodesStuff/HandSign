```python
cap = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Check if frame is captured
    if ret:
        # Get the dimensions of the frame
        height, width, _ = frame.shape
        
        # Calculate the target height for a 5:3 aspect ratio
        target_height = int((5/3) * width)
        
        # Calculate the target width for a 5:3 aspect ratio
        target_width = int((3/5) * height)

        # Determine if we need to crop based on width or height
        if target_height <= height:
            # Center crop height
            start_y = (height - target_height) // 2
            frame_cropped = frame[start_y:start_y + target_height, :]
        elif target_width <= width:
            # Center crop width
            start_x = (width - target_width) // 2
            frame_cropped = frame[:, start_x:start_x + target_width]
        else:
            frame_cropped = frame
        
        # Display the resulting frame
        cv2.imshow('5:3 Aspect Ratio', frame_cropped)

        sentence, prev_prediction = process_frame(frame_cropped, labels_dict, sentence, prev_prediction)
        print(sentence, prev_prediction, counter)
        counter += 1

        
        if counter > 30:
            counter = 0
            sentence = " "
            break

        
        
        # Break loop on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        print("Failed to capture frame")

# Release the capture when done
cap.release()
cv2.destroyAllWindows()
```
